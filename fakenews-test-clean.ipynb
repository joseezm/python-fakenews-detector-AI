```python
import pandas as pd

from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from gensim.parsing.preprocessing import remove_stopwords
```

```python
data = pd.read_csv('dataset/test.csv')
labels = pd.read_csv('dataset/submit.csv')
data['label'] = labels['label']
data.head(20)
```

```python
# QUITAR FILAS CON ESPACIOS VACIOS
data = data.dropna(axis=0, how = "any")

# QUITAR COLUMNAS ID, TITULO y AUTOR
data = data.drop(['id','title','author'], axis=1)

# TRANSFORMAR EL TEXTO A MAYÚSCULAS
data['text'] = data['text'].str.lower()

data.head(20)
```

```python
text_wsw = []
for i in data['text']:
    text_wsw.append(remove_stopwords(i))
print(text_wsw[1])
```

```python
data['text'] = text_wsw
data.head(20)
```

```python
data['text'] = data['text'].str.replace(r'[^\w\s]', '')
data.head(20)
```

```python
print(len(data['text']))
data = data[data.text != '']
print(len(data['text']))
```
